```python
20
a = float(input("Enter value for a: "))
b = float(input("Enter value for b: "))

print("\nSelect operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")


choice = input("Enter choice (1/2/3/4): ")


match choice:
    case '1':
        print("Result:", a + b)
    case '2':
        print("Result:", a - b)
    case '3':
        print("Result:", a * b)
    case '4':
        if b == 0:
            print("Cannot divide by zero.")
        else:
            print("Result:", a / b)
    case _:
        print("Invalid choice.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
a=np.array([10,20,30,40])
b=np.array([20,40,60,80])
plt.xlabel("a value")
plt.ylabel("b values")
plt.title("graph of x vs y")
plt.plot(a,b)
plt.grid()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Line Graph x vs y")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ['Math', 'Science', 'English', 'History']
marks = [85, 90, 80, 70]

plt.bar(subjects, marks, color='red')
plt.title("Bar Graph of Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [45, 30, 15, 10]


plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("PIE CHART EXAMPLE")
plt.axis('equal')  
plt.show()
```
